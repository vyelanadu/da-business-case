```python
# Age min max (18, 50)
age_range = [17,25,35,45,51]
age_labels = ['Young', 'Adult', 'Middle Aged', 'Elder']
df['age_category'] = pd.cut(df['Age'],bins = age_range,labels = age_labels)

# Income min and max (29562, 104581)
income_range = [29561,40000,60000,80000,104582]
income_labels = ['Low','Moderate','High','Very High']
df['income_category'] = pd.cut(df['Income'],bins = income_range,labels = income_labels)

# Miles min and max (21, 360)
miles_range = [20,50,100,200,361]
miles_label = ['Less Active', 'Moderately Active', 'Active', 'Fitness Enthusiast']
df['miles_category'] = pd.cut(df['Miles'],bins = miles_range,labels = miles_label)
```

```python
df.isnull().sum()
```

```python
pd.crosstab(index =df['Product'],columns = df['age_category'],margins = True,normalize = True).round(2)
```

## Observation
Young and Adult people prefer buying KP281 and KP481

```python
pd.crosstab(index =df['Product'],columns = df['income_category'],margins = True,normalize = True ).round(2)
```

## Observation 
Income category "Low" and "Moderate" prefer buying KP281.

Income category "High" and "Very High" prefer buying KP781.

```python
pd.crosstab(index =df['Product'],columns = df['miles_category'],margins = True,normalize = True ).round(2)
```

## Observation
Active and Fitness Enthusiast prefer KP781.


## Suggestion
Most product buyers age are around 20-30. By Marketing to these age group can help in increase sales.

Moderate income group (40000,60000) tent to buy product KP281 and KP481. Attractive discount and easy payment option can help in increase in sales of KP781 in this income group

Buyers with fitness rating of 5 tend to buy top varient of the product KP781.

Product KP781 sales among young and adult category (18-35) are very less.Target these age group for upsell.
